```
import requests
from bs4 import BeautifulSoup
 
 
 
r = requests.get('https://careerbuilder.vn/viec-lam/data-scientist-k-vi.html');
 
html_text = r.text
tree = BeautifulSoup(html_text, 'html.parser')
```

```
jobs= tree.find_all('div',{'class':'job-item'})
```

```
data={'title':[],'url':[],'company':[],'salary':[],'location':[],'posted_date':[],'detail':[]};

for job in jobs:


  title = job.find('a', {'class': 'job_link'})['title'] # Nếu dùng .string.strip() thì có một tuyển dụng bị lỗi
  data['title'].append(title);
  
  url = job.find('a', {'class': 'job_link'})['href'] 
  data['url'].append(url);

  company = job.find('a', {'class': 'company-name'}).string
  data['company'].append(company);

  salary = job.find('div', {'class': 'salary'}).text
  data['salary'].append(salary);

  location = job.find('div', {'class': 'location'}).text.strip()
  data['location'].append(location);

  posted_date = job.find('time').string
  data['posted_date'].append(posted_date);

  r = requests.get(url);
  html_text = r.text;
  tree = BeautifulSoup(html_text, 'html.parser')
  details=tree.find_all('div',{'class':'detail-row'})
  s="";
  for detail in details:
    s=s+detail.text.strip();

  data['detail'].append(s);

print('title')
```

```
import pandas as pd

df=pd.DataFrame(data,columns=['title','url','company','salary','location','posted_date','detail']);

df.to_csv('/content/drive/My Drive/Môn học/Introduce Data Science/WebCrawling1/data.csv')
```
